```python
import pandas as pd

# Učitavanje podataka
df = pd.read_csv("student_mental_health_burnout.csv")

# Prvih 5 redaka
print("Prvih 5 redaka \n:", df.head())
```

```python
# Veličina skupa (broj redaka, broj stupaca)
print("\n2. Velicina skupa podataka:", df.shape)

```

```python
# Nazivi stupaca
print("\n3. Nazivi stupaca:", df.columns)
```

```python
# Nedostajuće vrijednosti
print("\n4. Broj nedostajucih vrijednosti po stupcu:\n",df.isna().sum())
```

```python
# Jedinstvene vrijednosti po stupcu
print("\n5. Jedinstvene vrijednosti po stupcu:")
for column in df.columns:
    print(f"{column}: {df[column].unique()}")
```

```python
# Tipovi podataka
print("\n6. Tipovi podataka po stupcu:\n", df.dtypes)
```

```python
# Frekvencije vrijednosti po stupcu
print("\n7. Frekvencije vrijednosti po stupcu:")
for column in df.columns:
    print(f"{column}:\n", df[column].value_counts())
    print("\n")
```
